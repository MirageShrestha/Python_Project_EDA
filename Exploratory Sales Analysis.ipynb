```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualizing data
%matplotlib inline
import seaborn as sns
```

```python
#DATA CLEANING
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
```

```python
df.drop(['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
#EXPLORATORY DATA ANALYSIS
```

```python
df.columns
```

```python
ax=sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
# From above graph we can see that most of the buyers are females and even the purchasing power of females are greater than males.
```

```python
ax = sns.countplot(data = df, x='Age Group')
```

```python
ax = sns.countplot(data = df, x='Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#TOTAL AMOUNT Vs AGE GROUP
```

```python
sales_age=df.groupby(['Age Group'], as_index=False).sum().sort_values(by='Amount', ascending=False)
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#From above graph we can see that most of the buyers are from the age group of 26-35 years female
```

```python
#STATE
```

```python
df.columns
```

```python
df.info()
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state, x ='State', y='Orders')
```

```python
sales_amt = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_amt, x ='State', y='Amount')
```

```python
#From above graphs we can see that unexpectedly most of the orders are from UP, Maharastra, and Krrnatka resp.
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize': (7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue ='Gender')
```

```python
#From above graphs we can see that most of the buyers are married(women) and they have high purchasing power
```

```python
#OCCUPATION
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
state_sales=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation', y='Amount', data=state_sales)
```

```python
#From above we can see that the most of the buyers are working in IT, Healthcare and Aviation
```

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(30,7)})
ax=sns.countplot(x='Product_Category', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
state_sales=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,7)})
sns.barplot(x='Product_Category', y='Amount', data=state_sales)
```

```python
#From above chart we can see that most of the sold products are from Food, Clothing, Footwear and Electronic Category
```

```python
sales_state=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.barplot(x='Product_ID', y='Orders', data=sales_state)
for bars in ax.containers:
    ax.bar_label(bars)
```
